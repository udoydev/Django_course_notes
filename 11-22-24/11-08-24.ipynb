```python
marks = int (input("enter your marks : "))
if marks>0 and marks<101:
    if ((marks>=40 and marks<50)):
        print("passed and you got D")
    elif ((marks>=50 and marks<60)):
        print("passed and you got C")
    elif ((marks>=60 and marks<70)):
        print("passed and you got B")
    elif ((marks>=70 and marks<80)):
        print("passed and you got A")
    elif ((marks>=80 and marks<=100)):
        print("passed and you got A+")
    else:
        print("Failed and you got F")
else:
    print("Invalid mark maybe you typed negative or greater than 100")
```
